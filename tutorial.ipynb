*All pre-trained models expect input images normalized in the same way, i.e. mini-batches of 3-channel RGB images of shape (3 x H x W), where H and W are expected to be at least 224. The images have to be loaded in to a range of [0, 1] and then normalized using mean = [0.485, 0.456, 0.406] and std = [0.229, 0.224, 0.225]. You can use the following transform to normalize:*

*normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])*

```python
# here we load our dataset
# taken from https://pytorch.org/tutorials/beginner/transfer_learning_tutorial.html
# Data augmentation and normalization for training Just normalization for validation
data_validation_transformers = transforms.Compose(
    [
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

data_dir = './images'
image_validation_dataset = datasets.ImageFolder(data_dir, data_validation_transformers)

validation_dataloader = torch.utils.data.DataLoader(image_validation_dataset, 
                                                    batch_size=4, num_workers=4, shuffle=True)

class_names = image_validation_dataset.classes
print('our dataset categories: ', class_names)

# we check if CUDA and GPU is available
# if You don't have GPU and CUDA You can try to find some free server which support it ? I will list some 
# free available resources at the end of the notebook.
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print('device type: ', device)
```

resnet18 was created with usage of ImageNet and a category which we are looking for is http://image-net.org/search?q=bear **bear**, we need to use the same name for a images directory so it is used as class name for images.

```python
# https://pytorch.org/tutorials/beginner/transfer_learning_tutorial.html
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(validation_dataloader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

**we skip training the model for now as we want to check first how our dataset will be evaluated by pre-trained resnet18**

```python
# we want to check how well model will work with new dataset withouth any changes
# so we load original class names
# https://github.com/ageron/handson-ml/blob/master/datasets/inception/imagenet_class_names.txt
# although not sure if above the class names list is correct

!wget https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/inception/imagenet_class_names.txt
with open('./imagenet_class_names.txt') as f:
    class_names = [line.strip() for line in f.readlines()]
```

```python
# https://pytorch.org/tutorials/beginner/transfer_learning_tutorial.html#visualizing-the-model-predictions
def visualize_model(model, num_images=6):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(validation_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title('predicted: {}'.format(class_names[preds[j]]))
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
visualize_model(resnet18)
```

**Based on above we see that prediction was correct.**
